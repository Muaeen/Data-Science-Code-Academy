```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
data = pd.read_csv('population2.csv')
data.head()
```

```python
data['Country Code'].unique()
```

```python
data[data['Country Name'].str.startswith('U',na=False)]
```

```python
Oman = data[data['Country Name'] == 'Oman'].reset_index(drop=True)
UK = data[data['Country Code'] == 'GBR'].reset_index(drop=True)
USA = data[data['Country Code'] == 'USA'].reset_index(drop=True)
```

```python
UK
```

```python
Oman.drop(columns=(['Country Code','Indicator Name','Indicator Code']),inplace=True)
UK.drop(columns=(['Country Code','Indicator Name','Indicator Code']),inplace=True)
USA.drop(columns=(['Country Code','Indicator Name','Indicator Code']),inplace=True)
```

```python
Oman = Oman.melt(id_vars='Country Name',var_name='Year',value_name='Population')
UK = UK.melt(id_vars='Country Name',var_name='Year',value_name='Population')
USA = USA.melt(id_vars='Country Name',var_name='Year',value_name='Population')
```

```python
Oman
```

```python
Oman['Population'].mean()
```

```python
plt.figure(figsize=(12,8))
plt.hist(Oman['Population'],bins=30,edgecolor='black')
plt.title('Population Distribution from 1960 to 2021')
plt.xlabel('Population')
plt.ylabel('Frequency');
```

```python
plt.figure(figsize=(12,8))
plt.hist(UK['Population'],bins=30,edgecolor='black')
plt.title('Population Distribution from 1960 to 2021')
plt.xlabel('Population')
plt.ylabel('Frequency');
```

```python
plt.figure(figsize=(12,8))
plt.hist(USA['Population'],bins=30,edgecolor='black')
plt.title('Population Distrbution from 1960 to 2021')
plt.xlabel('Population')
plt.ylabel('Frequency');
```

```python
fig,ax = plt.subplots()
ax.bar(Oman['Year'],Oman['Population'])
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
plt.xticks(rotation=45)
plt.title('Oman Population from 1960 to 2021')
plt.xlabel('Year')
plt.ylabel('Population');
```

```python
fig,ax = plt.subplots()
plt.bar(UK['Year'],UK['Population'])
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
plt.xticks(rotation=45)
plt.title('UK Population from 1960 to 2021')
plt.xlabel('Year')
plt.ylabel('Population');
```

```python
fig,ax = plt.subplots()
plt.bar(USA['Year'],USA['Population'])
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
plt.xticks(rotation=45)
plt.title('USA population from 1960 to 2021')
plt.xlabel('Year')
plt.ylabel('Population');
```
